## Courtesy example from - https://deallen7.medium.com/how-to-create-a-pandas-dataframe-from-an-api-endpoint-in-a-jupyter-notebook-f2561f766ca3

```python
import pandas as pd
import requests
```

```python
url = 'https://api.covid19api.com/summary'
```

```python
r = requests.get(url)
```

```python
json = r.json()
```
